# Correlation

# Observations:

There is a strong relationship between energy and loundness of the tracks.

There is a significant but not as strong, relationship between valence and dancibility of the tracks.

```python
dftc = dft.drop(columns = ["key","mode","explicit"]).corr()

heatmap=sns.heatmap(dftc,annot=True,fmt='.1g',cmap="inferno",linewidth=1)
plt.figure(figsize=(140,60))
heatmap.set_title("Correlation Map")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)

plt.show()
```

# Sample data to conduct regression analysis

```python
sample_dft=dft.sample(int(0.005*len(dft)))
len(sample_dft)
```

# Observation:

Loudness and Energy of a track have a strong positive (correlation coefficient being 0.8) relationship, and an increase/decrease in one of the either, implies the same on the other, i.e., an increase in loudness would be associated with an increase in energy, and vice versa.

```python
plt.figure(figsize=(10,6))
sns.regplot(sample_dft,x='energy',y='loudness',color='pink').set(title="Loudness vs Energy")
```

# Observation:

Acousticness and Popularity of a track have a negative (correlation coefficient being -0.4) relationship, and an increase/decrease in one of the either, implies the opposite on the other, i.e., an increase in acousticness would be associated with a decrease in popularity, and vice versa.

```python
plt.figure(figsize=(10,6))
sns.regplot(sample_dft,x='acousticness',y='popularity',color='c').set(title="Acousticness vs Popularity")
```

# Extracting "Year"of release to plot a "HISTOGRAM"

```python
year=dft.index.year
year
```

```python
dft["year"]=year
```

# Observation:

Then number of releases has rapidly increased from early 1920s, showing a major dip in late 1960s, early 1990s and late 1990s. A major dip is observed in the begining of early 2020s, due to the impact of the COVID-19 pandemic. 

```python
sns.displot(dft['year'],discrete=True,aspect=True,height=5,kind='hist',color='g')

plt.show()
```

# Observation:

The duration of the tracks has, on an average, been between 160 to 240 seconds. This specific range can be clearly seen in the linechart, displayed exactly below the barchart. 

```python
plt.figure(figsize=(15,5))
plt.title("Year vs Time-in seconds")
ax=sns.barplot(x='year',y='seconds',data=dft,errwidth=False)
ax.tick_params(axis='x',rotation=90)

plt.show()
```

```python
plt.figure(figsize=(14,7))
plt.title("Year vs Time-in seconds")
sns.set_style(style='whitegrid')
sns.lineplot(x='year',y='seconds',data=dft)
plt.xticks(rotation=60)

plt.show()
```

# Observation:

"Classical" and "World" genres, have the maximum duration tracks. 

Whereas the duration of "Children's Music" tracks is the shortest on an average.

```python
plt.title("Time as per Genre")
sns.barplot(x='duration_ms',y='genre',data=dff)

plt.show()
```
